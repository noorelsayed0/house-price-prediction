## feature engineering

```python
df.columns
```

```python
df=df.drop('id',axis=1)
```

```python
df['date']=pd.to_datetime(df['date'])  #convert from object  data type to datetime data type
df['date']
```

```python
df['year']=df['date'].apply(lambda x:x.year) 
df['month']=df['date'].apply(lambda x:x.month) 
df['day']=df['date'].apply(lambda x:x.day) 
```

```python
plt.figure(figsize=(12,5))
df.groupby('month').describe()['price']['mean'].plot()
```

```python
plt.figure(figsize=(12,5))
df.groupby('year').describe()['price']['mean'].plot()
```

```python
df=df.drop('day',axis=1)
```

```python
df.columns
```

```python
df['zipcode'].value_counts()  
```

```python
df=df.drop('zipcode',axis=1) # bacause it will be hard to use get dummies to 70 unique values when converting this to string
```

```python
df['yr_renovated'].value_counts()
```

```python
df['yr_renovated'].apply(lambda x: 1 if x != 0 else 0).value_counts()  #or we could do nothing because as year increases price also increase
```

### split the data and scale it

```python
x=df.drop('price',axis=1).values
y=df['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=10)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler =MinMaxScaler()
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

### model training

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

```python
model=Sequential()
model.add(Dense(19,activation ='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(19,activation='relu'))
model.add(Dense(1))
model.compile(optimizer='adam',loss='mse')
```

```python
model.fit(x=X_train,y=y_train,
validation_data=(X_test,y_test),
batch_size=128,
epochs=400)
```

```python
pd.DataFrame(model.history.history).plot()
```

### evaluation

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error
```

```python
pred=model.predict(X_test)
```

```python
pred
```

```python
mean_squared_error(y_test,pred)
```

```python
np.sqrt(mean_squared_error(y_test,pred))
```

```python
mean_absolute_error(y_test,pred)
```
